## dividing the data set with K-fold

```python
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
```

```python
# the 10-fold
kfold = KFold(n_splits=10, shuffle=True, random_state=123)
for train, test in kfold.split(X):
    print(train, test)
```

```python
# training-testing function
def train_model_by_kfold(df, model):
    kfold = KFold(n_splits=10, shuffle=True, random_state=123)
    X = df.drop('target', 1).values
    y = df['target'].values
    
    res = pd.DataFrame()
    k = 0
    for train, test in kfold.split(X):
        k+=1
        model.fit(X[train], y[train])
        
        train_score = roc_auc_score(y[train], model.predict_proba(X[train])[:,1])
        valid_score = roc_auc_score(y[test], model.predict_proba(X[test])[:,1])
        res = res.append({'k': k,
                          'train_score': train_score, 
                          'valid_score': valid_score}, ignore_index=True)

    print('Training accuracy score: %.3f +/- %.3f\nValidation accuracy score %.3f +/- %.3f' % (res['train_score'].mean(), res['train_score'].std(), res['valid_score'].mean(), res['valid_score'].std()))
    return res
```

## ANN

```python
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# training-testing function with StandardScaler
def train_model_by_kfold_with_scaler(df, model):
    scaler = StandardScaler()

    kfold = KFold(n_splits=10, shuffle=True, random_state=123)
    X = df.drop('target', 1).values
    X = scaler.fit_transform(X)
    y = df['target'].values
    
    res = pd.DataFrame()
    k = 0
    for train, test in kfold.split(X):
        k+=1
        model.fit(X[train], y[train])
        
        train_score = roc_auc_score(y[train], model.predict_proba(X[train])[:,1])
        valid_score = roc_auc_score(y[test], model.predict_proba(X[test])[:,1])
        res = res.append({'k': k,
                          'train_score': train_score, 
                          'valid_score': valid_score}, ignore_index=True)

    print('Training accuracy score: %.3f +/- %.3f\nValidation accuracy score %.3f +/- %.3f' % (res['train_score'].mean(), res['train_score'].std(), res['valid_score'].mean(), res['valid_score'].std()))
    return res
```

```python
model = MLPClassifier()
res_ann_grid = train_model_by_kfold_with_scaler(df, model)
```

### Hyperparameter tuning

#### tunning hidden_layer_size (1 to 151, 50)

```python
scaler = StandardScaler()
kfold = KFold(n_splits=10, shuffle=True, random_state=123)
X = df.drop('target', 1).values
X = scaler.fit_transform(X)
y = df['target'].values
        
res = pd.DataFrame()
k=0

for train, test in kfold.split(X):
    k+=1
    for size_ in range(1, 200, 50):
        print(f"k: {k}, size: {size_}")
        model = MLPClassifier(hidden_layer_sizes=(size_, size_))
        model.fit(X[train], y[train])
        res = res.append({'k': k,
                          'hidden_layer_sizes': size_,
                          'train_acc': roc_auc_score(y[train], model.predict_proba(X[train])[:,1]),
                          'test_acc': roc_auc_score(y[test], model.predict_proba(X[test])[:,1])}, ignore_index=True)
```

```python
res.head()
```

```python
final_res = res[['hidden_layer_sizes', 'train_acc', 'test_acc']].groupby(['hidden_layer_sizes']).mean().reset_index().sort_values('test_acc', ascending=False)
final_res.head()
```

```python
chart_res = final_res.sort_values('hidden_layer_sizes')
plt.figure(figsize=(13, 4))
plt.plot(chart_res['hidden_layer_sizes'], chart_res['train_acc'], marker='o', markersize=4)
plt.plot(chart_res['hidden_layer_sizes'], chart_res['test_acc'], marker='o', markersize=4)
plt.legend(['Train accuracy', 'Test accuracy'])
plt.xlabel('# neurons')
plt.title('Tuning the number of neurons in 2 layers networks ')
plt.show()
```

```python
scaler = StandardScaler()
kfold = KFold(n_splits=10, shuffle=True, random_state=123)
X = df.drop('target', 1).values
X = scaler.fit_transform(X)
y = df['target'].values
        
res = pd.DataFrame()
k=0

active_list = ['logistic', 'tanh', 'relu']

for train, test in kfold.split(X):
    k+=1
    for func_ in active_list:
        print(f"k: {k}, func: {func_}")
        model = MLPClassifier(hidden_layer_sizes=(13, 13), max_iter=450, activation=func_)
        model.fit(X[train], y[train])
        res = res.append({'k': k,
                          'activation': func_,
                          'train_acc': roc_auc_score(y[train], model.predict_proba(X[train])[:,1]),
                          'test_acc': roc_auc_score(y[test], model.predict_proba(X[test])[:,1])}, ignore_index=True)
```

```python
# training-testing function with StandardScaler (descicon_function instead of predict_proba)
def train_model_by_kfold_with_scaler_for_SVM(df, model):
    scaler = StandardScaler()

    kfold = KFold(n_splits=10, shuffle=True, random_state=123)
    X = df.drop('target', 1).values
    X = scaler.fit_transform(X)
    # reducing number of vars to 2
    #     pca = PCA(n_components=2)
    #     X = pca.fit_transform(X)
    y = df['target'].values
    
    res = pd.DataFrame()
    k = 0
    for train, test in kfold.split(X):
        k+=1
        model.fit(X[train], y[train])
        
        train_score = roc_auc_score(y[train], model.decision_function(X[train]))
        valid_score = roc_auc_score(y[test], model.decision_function(X[test]))
        res = res.append({'k': k,
                          'train_score': train_score, 
                          'valid_score': valid_score}, ignore_index=True)

    print('Training accuracy score: %.3f +/- %.3f\nValidation accuracy score %.3f +/- %.3f' % (res['train_score'].mean(), res['train_score'].std(), res['valid_score'].mean(), res['valid_score'].std()))
    return res
```

## Unsupervised Learning - Clustering

### k medoids

```python
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
from pyclustering.utils.metric import distance_metric, type_metric
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.cluster.center_initializer import kmeans_plusplus_initializer
from sklearn.metrics import silhouette_score, davies_bouldin_score
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
# todo: change x below
# todo: download pyclustering package using the command (in the terminal) : pip install pyclustering
x = df.drop('target', 1).values #'your dataset without the target variable - (ndarray - numpy array in shape (samples, features))'

metric = distance_metric(type_metric.GOWER, max_range=x.max(axis=0))

dbi_list = []
sil_list = []

max_n_clusters = 8

# this part should take you about 40-60 minutes of calculations (maybe more - depends on your computer)
for n_clusters in tqdm(range(2, max_n_clusters, 1)):
    initial_medoids = kmeans_plusplus_initializer(x, n_clusters).initialize(return_index=True)
    kmedoids_instance = kmedoids(x, initial_medoids, metric=metric)
    kmedoids_instance.process()
    assignment = kmedoids_instance.predict(x)

    sil = silhouette_score(x, assignment)
    dbi = davies_bouldin_score(x, assignment)

    dbi_list.append(dbi)
    sil_list.append(sil)


plt.plot(range(2, max_n_clusters, 1), sil_list, marker='o')
plt.title("Silhouette")
plt.xlabel("Number of clusters")
plt.show()

plt.plot(range(2, max_n_clusters, 1), dbi_list, marker='o')
plt.title("Davies-bouldin")
plt.xlabel("Number of clusters")
plt.show()
```

```python
## pre-processing

### handling nulls

#### gender

new_test = test.copy()
new_test['gender'].fillna('Unknown', inplace=True)

#### enrolled_university

new_test['enrolled_university'].fillna('no_enrollment', inplace=True)

#### education_level

new_test[new_test['education_level'].notnull() == False]

edc_list = np.array_split(new_test[new_test['education_level'].notnull() == False], 5)

edc_list[0]['education_level'].fillna("Phd", inplace=True)
edc_list[1]['education_level'].fillna("Masters", inplace=True)
edc_list[2]['education_level'].fillna("Graduate", inplace=True)
edc_list[3]['education_level'].fillna("High School", inplace=True)
edc_list[4]['education_level'].fillna("Primary School", inplace=True)

result = pd.concat(edc_list)

frames = [result, new_test[new_test['education_level'].notnull() == True]]
temp = pd.concat(frames)

new_test = temp

#### major_discipline

new_test['major_discipline'].fillna('STEM', inplace=True)

#### company_size & company_type

# populate the popular size for each type
new_test.loc[(new_test.company_type == 'Pvt Ltd') & (new_test.company_size.notnull() == False),'company_size']='50-99'
new_test.loc[(new_test.company_type == 'Public Sector') & (new_test.company_size.notnull() == False),'company_size']='1000-4999'
new_test.loc[(new_test.company_type == 'Other') & (new_test.company_size.notnull() == False),'company_size']='100-500'
new_test.loc[(new_test.company_type == 'NGO') & (new_test.company_size.notnull() == False),'company_size']='100-500'
new_test.loc[(new_test.company_type == 'Funded Startup') & (new_test.company_size.notnull() == False),'company_size']='50-99'
new_test.loc[(new_test.company_type == 'Early Stage Startup') & (new_test.company_size.notnull() == False),'company_size']='<10'

# populate the popular type for each size
new_test.loc[(new_test.company_size.notnull() == True) & (new_test.company_type.notnull() == False),'company_type']='Pvt Ltd'

# fill the rest as Unknown
new_test['company_type'].fillna("Unknown", inplace=True)
new_test['company_size'].fillna("Unknown", inplace=True)

#### last_new_job

new_test['last_new_job'].fillna(1, inplace=True)

#### experience

new_test['experience'].fillna(0, inplace=True)
new_test['experience'].replace(">20", 21, inplace=True)
new_test['experience'].replace("<1", 0.5, inplace=True)

## feature extraction

### age

def age_by_edu(edu_level):
    if edu_level == "Primary School":
        return 15
    elif edu_level == "High School":
        return 18
    elif edu_level == "Graduate":
        return 22
    elif edu_level == "Masters":
        return 27
    else:
        return 31

new_test['age_by_edu'] = new_test['education_level'].apply(age_by_edu)

new_test['age'] = new_test['age_by_edu']+new_test['experience']

### is_first_job

new_test['is_first_job'] = new_test.apply(lambda row: 1 if ((row['last_new_job'] == "never") and (row["relevent_experience"] == "No relevent experience")) else 0, axis=1)

## feature representation

#### city_development_index_norm

new_test['city_development_index_norm'] = new_test['city_development_index']/new_test['city_development_index'].max()

#### relevent_experience_bool

new_test['relevent_experience_bool'] = new_test['relevent_experience'].apply(lambda row: 1 if row=="Has relevent experience" else 0)

#### experience

new_test['experience_norm'] = new_test['experience']/new_test['experience'].max()

#### company_type

def is_pvt_ltd(x):
    if x == "Pvt Ltd":
        return 1
    else:
        return 0

new_test['is_pvt_ltd'] = new_test['company_type'].apply(is_pvt_ltd)

def is_type_unknown(x):
    if x == "Unknown":
        return 1
    else:
        return 0

new_test['is_type_unknown'] = new_test['company_type'].apply(is_type_unknown)

def is_funded_startup(x):
    if x == "Funded Startup":
        return 1
    else:
        return 0

new_test['is_funded_startup'] = new_test['company_type'].apply(is_funded_startup)

def is_public_sector(x):
    if x == "Public Sector":
        return 1
    else:
        return 0

new_test['is_public_sector'] = new_test['company_type'].apply(is_public_sector)

def is_early_stage_startup(x):
    if x == "Early Stage Startup":
        return 1
    else:
        return 0

new_test['is_early_stage_startup'] = new_test['company_type'].apply(is_early_stage_startup)

def is_NGO(x):
    if x == "NGO":
        return 1
    else:
        return 0

new_test['is_NGO'] = new_test['company_type'].apply(is_NGO)

def is_other(x):
    if x == "Other":
        return 1
    else:
        return 0

new_test['is_other'] = new_test['company_type'].apply(is_other)

#### last_new_job

def last_new_job_int(x):
    if x == 1:
        return 1
    elif x == 2:
        return 2
    elif x == 3:
        return 3
    elif x == 4:
        return 4
    elif x == ">4":
        return 5
    else:
        return 0

new_test['last_new_job_int'] = new_test['last_new_job'].apply(last_new_job_int)
new_test['last_new_job_norm'] = new_test['last_new_job_int']/new_test['last_new_job_int'].max()

#### enrolled_university

def enrolled_university_norm(x):
    if x == "no_enrollment":
        return 0
    elif x == "Part time course":
        return 0.5
    else:
        return 1
    
new_test['enrolled_university_norm'] = new_test['enrolled_university'].apply(enrolled_university_norm)

#### major_discipline

new_test['major_discipline'].value_counts()

new_test = pd.get_dummies(new_test, columns=['major_discipline'])

#### education_level

def education_level_norm(x):
    if x == "Primary School":
        return 0
    elif x == "High School":
        return 0.25
    elif x == "Graduate":
        return 0.5
    elif x == "Masters":
        return 0.75
    else:
        return 1
    
new_test['education_level_norm'] = new_test['education_level'].apply(education_level_norm)

#### company_size

def company_size_norm(x):
    if x == "Unknown":
        return 0
    elif x == "<10":
        return 1
    elif x == "10-49":
        return 2
    elif x == "50-99":
        return 3
    elif x == "100-500":
        return 4
    elif x == "500-999":
        return 5
    elif x == "1000-4999":
        return 6
    elif x == "5000-9999":
        return 7
    else:
        return 8
    
new_test['company_size_int'] = new_test['company_size'].apply(company_size_norm)
new_test['company_size_norm'] = new_test['company_size_int']/new_test['company_size_int'].max()

#### last_new_job

new_test['last_new_job'].value_counts()

def last_new_job_int(x):
    if x == "never":
        return 0
    elif x == ">4":
        return 5
    else:
        return x
    
new_test['last_new_job_int'] = new_test['last_new_job'].apply(last_new_job_int)
new_test['last_new_job_norm'] = new_test['last_new_job_int']/new_test['last_new_job_int'].max()

#### training_hours

new_test['training_hours_norm'] = new_test['training_hours']/new_test['training_hours'].max()

new_test.head()
```
