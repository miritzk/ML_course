```python
def age_by_edu(edu_level):
    if edu_level == "Primary School":
        return 15
    elif edu_level == "High School":
        return 18
    elif edu_level == "Graduate":
        return 22
    elif edu_level == "Masters":
        return 27
    else:
        return 31

new_train['age_by_edu'] = new_train['education_level'].apply(age_by_edu)
new_train.head()
```

```python
new_train['age'] = new_train['age_by_edu']+new_train['experience']
new_train.head()
```

## dividing the data set with K-fold

```python
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
```

```python
# the 10-fold
kfold = KFold(n_splits=10, shuffle=True, random_state=123)
for train, test in kfold.split(X):
    print(train, test)
```

```python
# training-testing function
def train_model_by_kfold(df, model):
    kfold = KFold(n_splits=10, shuffle=True, random_state=123)
    X = df.drop('target', 1).values
    y = df['target'].values
    
    res = pd.DataFrame()
    k = 0
    for train, test in kfold.split(X):
        k+=1
        model.fit(X[train], y[train])
        
        train_score = roc_auc_score(y[train], model.predict_proba(X[train])[:,1])
        valid_score = roc_auc_score(y[test], model.predict_proba(X[test])[:,1])
        res = res.append({'k': k,
                          'train_score': train_score, 
                          'valid_score': valid_score}, ignore_index=True)

    print('Training accuracy score: %.3f +/- %.3f\nValidation accuracy score %.3f +/- %.3f' % (res['train_score'].mean(), res['train_score'].std(), res['valid_score'].mean(), res['valid_score'].std()))
    return res
```

### looking at max_depth with *holdout*

```python
max_depth_list = np.arange(1, 30, 1)
print(max_depth_list)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
res = pd.DataFrame()
for max_depth in max_depth_list:
    model = DecisionTreeClassifier(max_depth=max_depth, criterion='entropy', splitter='best')
    model.fit(X_train, y_train)
    res = res.append({'max_depth': max_depth,
                      'train_acc': roc_auc_score(y_train, model.predict_proba(X_train)[:,1]),
                      'test_acc': roc_auc_score(y_test, model.predict_proba(X_test)[:,1])}, ignore_index=True)
```

```python
plt.figure(figsize=(13, 4))
plt.plot(res['max_depth'], res['train_acc'], marker='o', markersize=4)
plt.plot(res['max_depth'], res['test_acc'], marker='o', markersize=4)
plt.legend(['Train accuracy', 'Test accuracy'])
plt.show()
```

### looking at max_depth with *k-fold*

```python
max_depth_list = np.arange(1, 30, 1)
res = pd.DataFrame()
k=0

for train, test in kfold.split(X):
    k+=1
    for max_depth in max_depth_list:
        model = DecisionTreeClassifier(max_depth=max_depth, criterion='entropy', splitter='best')
        model.fit(X[train], y[train])
        res = res.append({'k': k,
                          'max_depth': max_depth,
                          'train_acc': roc_auc_score(y[train], model.predict_proba(X[train])[:,1]),
                          'test_acc': roc_auc_score(y[test], model.predict_proba(X[test])[:,1])}, ignore_index=True)
```

```python
res.head()
```

```python
# max_depth = 5 had the best test accuracy (78.7%)
final_res = res[['max_depth', 'train_acc', 'test_acc']].groupby(['max_depth']).mean().reset_index().sort_values('test_acc', ascending=False)
final_res.head()
```

```python
chart_res = final_res.sort_values('max_depth')
plt.figure(figsize=(13, 4))
plt.plot(chart_res['max_depth'], chart_res['train_acc'], marker='o', markersize=4)
plt.plot(chart_res['max_depth'], chart_res['test_acc'], marker='o', markersize=4)
plt.legend(['Train accuracy', 'Test accuracy'])
plt.show()
```
