#### city_development_index

```python
bins = pd.cut(new_train["city_development_index"], 10)
bins
```

```python
bins.value_counts()
```

```python
new_train = pd.merge(new_train, bins, left_index=True, right_index=True, suffixes=("", "_bucket"))
new_train.head()
```

#### experience

```python
def experience_buckets(x):
    if x <= 3:
        return "0-3"
    elif x <= 10:
        return "4-10"
    else:
        return "11+"

new_train['experience'].replace(">20", 21, inplace=True)
new_train['experience'].replace("<1", 0.5, inplace=True)

new_train['experience_bucket'] = new_train['experience'].apply(experience_buckets)
new_train
```

#### training_hours

```python
def training_hours_buckets(x):
    if x <= 100:
        return "0-100"
    elif x <= 200:
        return "101-200"
    else:
        return "201+"

new_train['training_hours_bucket'] = new_train['training_hours'].apply(training_hours_buckets)
new_train
```

```python
def convert_city_index_bucket_to_int(x):
    if x == '(0.447, 0.498]':
        return 1
    elif x == '(0.498, 0.548]':
        return 2
    elif x == '(0.548, 0.598]':
        return 3
    elif x == '(0.598, 0.648]':
        return 4
    elif x == '(0.648, 0.698]':
        return 5
    elif x == '(0.698, 0.749]':
        return 6
    elif x == '(0.749, 0.799]':
        return 7
    elif x == '(0.799, 0.849]':
        return 8
    elif x == '(0.849, 0.899]':
        return 9
    else:
        return 10

new_train['city_development_index_rank'] = new_train['city_development_index_bucket_str'].apply(convert_city_index_bucket_to_int)
new_train.head()
```
